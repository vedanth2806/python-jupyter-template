### Exercise 4: Pie Chart with Percentages

```python
# TASK: Create a pie chart with the following data: labels = ['Python', 'Java', 'C++', 'JavaScript'] and sizes = [40, 25, 20, 15].
# Display the percentages on the chart using autopct.
```

```python
import matplotlib.pyplot as plt

# Step 1: Define the data for the pie chart
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

# Step 2: Create the pie chart with percentages
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)

# Step 3: Set the aspect ratio to ensure the pie is circular
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Step 4: Display the pie chart
plt.title('Programming Languages Distribution')
plt.show()
```
