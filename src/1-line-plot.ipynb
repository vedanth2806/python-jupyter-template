### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.// Import necessary libraries
import matplotlib.pyplot as plt
import numpy as np

# Step 1: Create x values from 0 to 10
x = np.arange(0, 11, 1)

# Step 2: Compute y values as the square of x
y = np.square(x)

# Step 3: Create the line plot
plt.plot(x, y, marker='o', linestyle='-', color='b')

# Step 4: Customize the plot
plt.title('Line Plot of y = x^2')
plt.xlabel('x values')
plt.ylabel('y values (x squared)')
plt.grid(True)

# Step 5: Display the plot
plt.show()
```
