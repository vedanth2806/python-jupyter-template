### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.

import matplotlib.pyplot as plt
import numpy as np

# Step 1: Create a 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Step 2: Plot a line chart in the first subplot
x = np.arange(0, 10)
y = x ** 2
axs[0, 0].plot(x, y, color='blue')
axs[0, 0].set_title('Line Plot')

# Step 3: Plot a bar chart in the second subplot
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
axs[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'purple'])
axs[0, 1].set_title('Bar Plot')

# Step 4: Plot a scatter plot in the third subplot
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]
axs[1, 0].scatter(x_scatter, y_scatter, color='orange')
axs[1, 0].set_title('Scatter Plot')

# Step 5: Plot a pie chart in the fourth subplot
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
axs[1, 1].set_title('Pie Chart')

# Step 6: Adjust layout and display the plot
plt.tight_layout()
plt.show()
```
